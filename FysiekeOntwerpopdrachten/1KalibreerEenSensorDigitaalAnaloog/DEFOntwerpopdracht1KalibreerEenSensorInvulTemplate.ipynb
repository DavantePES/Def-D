```python
# Zet hier jouw data in een voor jou handig formaat
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
# Data, bits 
controlemetinglux = [0] # 0 Lux correspondeert met 4 bits
controlemetingbits = [4] # 0 Lux correspondeert met 4 bits
lux = np.array([29, 28.9, 28.3, 29.4, 27.4, 7800, 7790, 7800, 7810, 7800, 680, 678, 679, 681, 680, 2540, 2530, 2540, 2550, 2530, 260, 259, 258, 260, 261, 1050, 1060, 1040, 1030, 1050, 1590, 1590, 1590, 1590, 1590, 4400, 4380, 4360, 4370, 4390, 440, 442, 441, 439, 438, 122, 121, 120, 121, 119, 52, 51.8, 52, 52.2, 51.5]) - controlemetinglux
bits = np.array([42, 37, 41, 39, 40, 720, 728, 730, 725, 724, 280, 280, 280, 290, 300, 510, 530, 540, 530, 530, 158, 155, 150, 148, 152, 360, 358, 355, 362, 366, 410, 420, 425, 422, 421, 625, 628, 640, 635, 630, 221, 223, 220, 219, 218, 112, 113, 114, 111, 110, 75, 77, 78, 79, 75]) - controlemetingbits
percentage = np.array([0, 0, 0, 0, 0, 100, 100, 100, 100, 100, 50, 50, 50, 50, 50, 80, 80, 80, 80, 80, 30, 30, 30, 30, 30, 60, 60, 60, 60, 60, 70, 70, 70, 70, 70, 80, 80, 80, 80, 80, 40, 40, 40, 40, 40, 20, 20, 20, 20, 20, 10, 10, 10, 10, 10])
averages_bits = np.array([np.mean(bits[0:5]), np.mean(bits[5:10]), np.mean(bits[10:15]), np.mean(bits[15:20]), np.mean(bits[20:25]), np.mean(bits[25:30]), np.mean(bits[30:35]), np.mean(bits[35:40]), np.mean(bits[40:45]), np.mean(bits[45:50])])
averages_lux = np.array([np.mean(lux[0:5]), np.mean(lux[5:10]), np.mean(lux[10:15]), np.mean(lux[15:20]), np.mean(lux[20:25]), np.mean(lux[25:30]), np.mean(lux[30:35]), np.mean(lux[35:40]), np.mean(lux[40:45]), np.mean(lux[45:50])])
```

```python
# Plot hier de meting en fit een lijn door de metingen.
# Maak gebruik van wat je geleerd hebt bij het practicum.
%matplotlib inline
averages_bits = np.sort(averages_bits)
averages_lux = np.sort(averages_lux)
plt.figure()
plt.plot(bits, lux, 'o', label='Data')
plt.plot(averages_bits, averages_lux, label='Gemiddelden')

def exp_model(x, a, n):
    return a * x ** n

x_fit = np.linspace(0, 800, 100)

popt, pcov = curve_fit(exp_model, averages_bits, averages_lux)

plt.plot(x_fit, exp_model(x_fit, *popt), label='Fit')
plt.legend()
plt.xlim(0, 800)
plt.ylim(0, 8000)
plt.show()

```
