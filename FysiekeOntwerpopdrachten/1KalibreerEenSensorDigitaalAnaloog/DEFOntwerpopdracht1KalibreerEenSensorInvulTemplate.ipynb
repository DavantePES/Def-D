```python
# Zet hier jouw data in een voor jou handig formaat
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
# Data, bits 
controlemetinglux = [0] # 0 Lux correspondeert met 4 bits
controlemetingbits = [4] # 0 Lux correspondeert met 4 bits
lux = np.array([29, 28.9, 28.3, 29.4, 27.4, 7800, 7790, 7800, 7810, 7800, 680, 678, 679, 681, 680, 2540, 2530, 2540, 2550, 2530, 260, 259, 258, 260, 261, 1040, 1030, 1040, 1030, 1050, 1590, 1590, 1590, 1590, 1590, 4400, 4380, 4410, 4390, 4400, 440, 442, 441, 439, 438, 122, 121, 120, 121, 119, 52, 51.8, 52, 52.2, 51.5]) - controlemetinglux # Metingen in lux min de controlemeting
bits = np.array([42, 37, 41, 39, 40, 720, 728, 730, 725, 724, 280, 280, 280, 290, 300, 510, 530, 540, 530, 530, 158, 155, 150, 148, 152, 360, 358, 355, 362, 366, 410, 420, 425, 422, 421, 625, 628, 640, 635, 630, 221, 223, 220, 219, 218, 112, 113, 114, 111, 110, 75, 77, 78, 79, 75]) - controlemetingbits # Metingen in bits min de controlemeting
percentage = np.array([0, 0, 0, 0, 0, 100, 100, 100, 100, 100, 50, 50, 50, 50, 50, 80, 80, 80, 80, 80, 30, 30, 30, 30, 30, 60, 60, 60, 60, 60, 70, 70, 70, 70, 70, 90, 90, 90, 90, 90, 40, 40, 40, 40, 40, 20, 20, 20, 20, 20, 10, 10, 10, 10, 10]) # Percentage van de schermhelderheid (afgelezen in settings van google pixel 7)
averages_bits = np.array([np.mean(bits[0:5]), np.mean(bits[5:10]), np.mean(bits[10:15]), np.mean(bits[15:20]), np.mean(bits[20:25]), np.mean(bits[25:30]), np.mean(bits[30:35]), np.mean(bits[35:40]), np.mean(bits[40:45]), np.mean(bits[45:50])])
averages_lux = np.array([np.mean(lux[0:5]), np.mean(lux[5:10]), np.mean(lux[10:15]), np.mean(lux[15:20]), np.mean(lux[20:25]), np.mean(lux[25:30]), np.mean(lux[30:35]), np.mean(lux[35:40]), np.mean(lux[40:45]), np.mean(lux[45:50])])
averages_percentage = np.array([0, 100, 50, 80, 30, 60, 70, 90, 40, 20])


# Plot hier de meting en fit een lijn door de metingen.
# Maak gebruik van wat je geleerd hebt bij het practicum.
%matplotlib inline
averages_bits = np.sort(averages_bits)
averages_lux = np.sort(averages_lux)

def exp_model(x, a, n, b):
    return a * x ** n + b

popt3, pcov3 = curve_fit(exp_model, np.sort(averages_percentage), np.sort(averages_bits))


# Van schermhelderheid in percentage naar bits

plt.plot(np.sort(averages_percentage), np.sort(averages_bits), 'ro', label='Data')
x_fit = np.linspace(0, 100, 100)
plt.plot(x_fit, exp_model(x_fit, *popt3), label='Fit')
plt.errorbar(np.sort(averages_percentage), np.sort(averages_bits), xerr=1, yerr=20, fmt='o', label='Foutmarge')
plt.plot(42, exp_model(42, *popt3), 'ko', label='Testmeting op 42%')
plt.legend()
plt.xlabel('Schermhelderheid (%)')
plt.ylabel('Bits')
plt.show()

# Testmeting
print("Testmeting op 42% gaf 226-230 bits, wat binnen de foutmarge van ", exp_model(42, *popt3), "bits +/- ", 0.05*exp_model(42, *popt3), "ligt.")
# onzekerheid in percentage
print(np.sqrt(np.diag(pcov3)) / popt3[0] * 100, "%")
```

```python
# van bits naar lux

averages_bits = np.sort(averages_bits)
averages_lux = np.sort(averages_lux)

plt.figure()

def exp_model(x, a, n, b):
    return a * x ** n + b

x_fit = np.linspace(0, 800, 100)

popt, pcov = curve_fit(exp_model, averages_bits, averages_lux)

plt.plot(x_fit, exp_model(x_fit, *popt), label='Fit')
plt.errorbar(x=averages_bits, y=averages_lux, yerr=0.2*averages_lux, xerr=20, fmt='o', label='Gemiddelden met foutbalken')
plt.legend()
plt.xlim(0, 800)
plt.ylim(0, 8000)
plt.show()
```
